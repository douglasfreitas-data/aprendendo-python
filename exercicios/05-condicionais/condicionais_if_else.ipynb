# Condicionais em Python

## if / else

Neste caso, dizemos qual instrução deve ser executada caso a primeira expressão lógica não seja verdadeira:

```python
if <expressão lógica>:
    <instrução>
else:
    <outra instrução>
```

````python
# Exemplo: verificar se um número é par ou ímpar
numero = 9

if numero % 2 == 0:
    print("Numero par")
else:
    print("Numero impar")

print("O programa continua")
````

É muito comum representarmos estas estruturas condicionais com fluxogramas. Nestes fluxogramas, fica bem fácil de perceber que realmente o fluxo do programa muda a depender se a expressão lógica é verdadeira ou falsa: 

````python
# Exemplo: bonificação de funcionários

funcionario = {
    "nome": "João",
    "idade": 25,
    "salario": 6000.0,
    "cargo": "Cientista de Dados",
    "tempo_de_casa": 3,
}

if funcionario["tempo_de_casa"] >= 5:
    funcionario["salario"] *= 1.2
    print(f"Parabéns {funcionario['nome']}, você recebeu um aumento de 20%")
else:
    print(f"Faltam {abs(funcionario['tempo_de_casa'] - 5)} anos para ser elegível ao bônus") 
````

````python
outro_funcionario = {
    "nome": "Maria",
    "idade": 30,
    "salario": 6500.0,
    "cargo": "Cientista de Dados",
    "tempo_de_casa": 6,
}

if outro_funcionario["tempo_de_casa"] >= 5:
    outro_funcionario["salario"] *= 1.2
    print(f"Parabéns {outro_funcionario['nome']}, você recebeu um aumento de 20%")
else:
    print(f"Faltam {abs(outro_funcionario['tempo_de_casa'] - 5)} anos para ser elegível ao bônus") 
````
