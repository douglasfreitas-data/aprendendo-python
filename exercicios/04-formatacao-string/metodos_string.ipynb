## Métodos de string

```python
exemplo = "curso de ciência de dados"
exemplo
```

```python
exemplo.capitalize() # Atribui a string a letra maiuscula a primeira letra
```

```python
exemplo.title()   # Atribui a letra maiuscula ao incio de todas as palavras
```

```python
exemplo_maisuculo = exemplo.upper()   #Atribui a letra maiscula a todas as letras
exemplo_maisuculo
```

```python
exemplo_maisuculo.lower()    #Atribui a letra minuscula a todas as letras
```

```python
exemplo.title().swapcase()  #Troca maiuscula por minuscula e minuscula por maiusculas
```

```python
exemplo.count("de")
```

```python
exemplo.split()   #Divide a string por espaços por padrão
```

```python
exemplo.split("de")   #Divide pela condição atribuida dentro de parentese
```

```python
exemplo.find("dados")   #Retorna a posição onde se inicia o parametro atribuido
```

```python
exemplo[20:-1]
```
