# Comparações

```python
salario_base = 1500
salario_minimo = 1621
```

```python
salario_base > salario_minimo #Resposta deve ser bool - boolean - boleano (true; false)
```

```python
salario_base == salario_minimo #Igual
```

```python
salario_base != salario_minimo #Diferente
```

```python
salario_base < salario_minimo == 1621 != salario_base
```

# Operador and

```python
salario_medio = 1730
```

```python
(salario_minimo < salario_medio) and (salario_minimo > salario_medio) #and - todas as condições devem ser atendidas para true
```

```python
(salario_minimo < salario_medio) and (salario_minimo < salario_medio) #and - todas as condições devem ser atendidas para true
```

# Operador or

```python
(salario_minimo < salario_medio) or (salario_minimo > salario_medio) #or - qualquer uma das condições atendidas para true
```

```python
(salario_minimo > salario_medio) and (salario_minimo > salario_medio) #or - nenhuma das condições atendidas para false
```
